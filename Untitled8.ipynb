```python
import sklearn
#from sklearn.datasets import load_boston
import numpy as np
import pandas as pd
import seaborn as sns

# Load the dataset
df = pd.read_csv('Groups.csv.zip')

# IQR
Q1 = np.percentile(df['DSA'], 25,
                interpolation = 'midpoint')

Q3 = np.percentile(df['DSA'], 75,
                interpolation = 'midpoint')
IQR = Q3 - Q1

print("Old Shape: ", df.shape)

# Upper bound
upper = np.where(df['DSA'] >= (Q3+1.5*IQR))

# Lower bound
lower = np.where(df['DSA'] <= (Q1-1.5*IQR))

# Removing the Outliers
df.drop(upper[0], inplace = True)
df.drop(lower[0], inplace = True)

print("New Shape: ", df.shape)

sns.boxplot(x='DSA', data=df)
```
